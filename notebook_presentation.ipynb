# Main Simulation: Educational virtual currency with RSA
In this notebook, we simulate the lifecycle of a cryptocurrency transaction using the classes defined in our project.

### 1. Imports & Initialization
We import:
*   `Wallet`: to manage users and keys.
*   `SimpleBlockchain` & `Transaction`: to manage the ledger and data.

We then initialize the blockchain and create our two participants, **Alice** and **Bob**.


```python
from wallet import Wallet 
from blockchain import SimpleBlockchain, Transaction

print("=== EDUCATIONAL VIRTUAL CURRENCY WITH RSA ===\n")

# 1. Create blockchain instance
blockchain = SimpleBlockchain()

# 2. Create two users wallets
print("1. Creating wallets...")
alice = Wallet("Alice")
bob = Wallet("Bob")
```

### 2. Identity Generation (RSA Keys)
Before any transaction can happen, users must generate their **RSA Key Pairs**:
*   **Private Key**: used to sign transactions (kept secret).
*   **Public Key**: used as the wallet address (shared with the network).

We then check their initial balance (which is 100 by default in the Wallet class).

```python
# 3. Generate RSA key pairs for both users
# These keys are used to sign transactions and verify identity
alice.generate_keys()
bob.generate_keys()

print(f"\n2. Alice has: {alice.balance} coin")
print(f"   Bob has: {bob.balance} coin")
```

### 3. Mining the first block
Alice decides to "mine" a block.
By calling `mine_pending_transactions`, her computer solves the cryptographic puzzle (Proof of Work).
*   **Reward**: as an incentive, the system rewards her with coins (simulated here by manually adding +10).
*   **Result**: a new block is added to the blockchain.

```python
# 4. Alice mines a block and receives a mining reward
# In a real blockchain, mining confirms pending transactions
print("\n3. Alice mines a block...")
blockchain.mine_pending_transactions(alice.get_public_key_pem())
alice.balance += 10 # Simplified mining reward simulation
```

### 4. The Transaction (Alice -> Bob)
This is the core security step.
1.  **Creation**: we define who sends (Alice), who receives (Bob), and how much (5 coins).
2.  **Signing**: Alice uses her **Private Key** to sign the transaction data. This generates a unique digital signature.
3.  **Verification**: the system uses Alice's **Public Key** to verify that the signature is valid and the data hasn't been tampered with.

If valid, the transaction is added to the "Pending Pool".

```python
# 5. Alice sends coins to Bob
print("\n4. Alice sends 5 coin to Bob...")

# Create transaction object
transaction = Transaction(
    alice.get_public_key_pem(),
    bob.get_public_key_pem(),
    5
)

# Prepare transaction data for signing
tx_data = {
    'sender': alice.get_public_key_pem(),
    'receiver': bob.get_public_key_pem(), 
    'amount': 5
}
# Alice signs the transaction with her private key
signature = alice.sign_transaction(tx_data, transaction.timestamp)

# Attach the signature to the transaction 
transaction.signature = signature

print(f"Signature created: {signature[:50]}...")
print(f"Is transaction valid? {transaction.is_valid()}")

# Add the transaction to the blockchain
if blockchain.add_transaction(transaction):
    # Update balances (simplified logic for demonstration)
    alice.balance -= 5
    bob.balance += 5
    print("Transaction accepted!")
else:
    print("Transaction rejected!")
```

### 5. Final Audit
Finally, we check:
1.  **Balances**: Alice should have less, Bob should have more.
2.  **Chain Integrity**: `is_chain_valid()` re-calculates all hashes to ensure no block was modified.
3.  **Ledger Display**: We print the full chain to see the blocks and their hashes.

```python
# 6. Show final balances
print(f"\n6. Final balances:")
print(f"   Alice: {alice.balance} coin")
print(f"   Bob: {bob.balance} coin")

# 7. Verify blockchain integrity
# Checks if hashes link correctly and if data has been tampered
print("\n7. Verifying blockchain integrity...")
blockchain.is_chain_valid()

# 8. Display the entire blockchain structure
print("\n8. Blockchain created:")
blockchain.display_chain()
```

### 6. Attack Simulation (Tampering Detection)
We simulate an attacker intercepting a valid transaction and modifying the amount from 5 coins to 10 coins.  
Because the transaction signature was generated for the original amount (5), changing the amount invalidates the signature.

```python
# 9. Simulate attack: attacker changes 5 -> 10
print("\n9. Simulating tampering attack...")

# Attacker copies original transaction
tampered_tx = Transaction(
    alice.get_public_key_pem(),
    bob.get_public_key_pem(),
    10  # maliciously changed from 5 to 10
)

# Reuse old signature (signature was for amount=5)
tampered_tx.signature = signature

print("Attacker modified transaction amount to 10")
print(f"Tampered transaction valid? {tampered_tx.is_valid()}")
```

The blockchain should detect this tampering and reject the malicious transaction.

```python
# Try adding malicious transaction
if blockchain.add_transaction(tampered_tx):
    print("ERROR: Tampered transaction was accepted!")
else:
    print("Attack detected. Tampered transaction rejected.")
```
